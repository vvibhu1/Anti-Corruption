```python
corperind = pd.read_excel('CorruptionPerceptionIndex-combined-2007to2019.xlsx')
corperind.head()
```

```python
corperind.drop('Unnamed: 6', axis = 1, inplace = True)
corperind.head()
```

```python
merged_cpi = pd.merge(df_usaid2, corperind,  how='left', left_on=["Country Code", "Year"], right_on = ['iso', 'Year'])
merged_cpi.head(50)
```

```python
merged_cpi.shape
```

```python
merged_cpi.drop(['country', 'iso', 'region'], axis = 1, inplace = True)
merged_cpi.head()
```

```python
merged_cpi['Corruption Perception Index'] = merged_cpi['scaled_score']
merged_cpi.drop('scaled_score', axis = 1, inplace = True)
merged_cpi.head()
```

```python
merged_cpi['Corruption Perception Rank'] = merged_cpi['rank']
merged_cpi.drop('rank', axis = 1, inplace = True)
```

```python
merged_cpi.to_csv('Step3_merged_ind_and_cpi_Jen.csv')
```
