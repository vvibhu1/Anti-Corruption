```python
df = pd.read_csv('USAID_Anticorruption_Projects_Database4.csv')
df
```

```python
# years and amount and corruption
df1 = df[["st_date", "awd_amount","corruption"]]
df1
```

```python
#clean the data(delete missing value)
df1.dropna()
```

```python
Corruption = {'High Corruption': 0,'Very High Corruption': 1} 
```

```python
df1.corruption = [Corruption[item] for item in df1.corruption] 
print(df1) 
```

```python
## classify investment into 3 level: 
#low: <1M,1M-5M, 5M-10M
#moderate: 10M-25M,25M-50M
#high:50M-100M,50M-100M,>100M
awd_amount = {'<1M':'Low Investment','1M-5M':'Low Investment','5M-10M': 'Moderate Investment','10M-25M': 'Moderate Investment','25M-50M': 'High Investment','50M-100M': 'High Investment','>100M': 'High Investment'} 
df1.awd_amount = [awd_amount[item] for item in df1.awd_amount] 
print(df1) 
```

```python
#group year
by_year = df1.groupby("st_date")["awd_amount"].count()
by_year
```

```python
levelofInvest=df1.to_csv("levelofInvest.csv")
```

```python
by_year1=df1.groupby(["st_date", "awd_amount"])["awd_amount"].count()
by_year1
```
