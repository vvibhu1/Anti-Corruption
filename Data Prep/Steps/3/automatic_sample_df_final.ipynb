```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
```

```python
df_education = pd.read_csv('df_ education.csv')
df_Macroeconomic_vulnerability_debt = pd.read_csv('df_ Macroeconomic_vulnerability_debt.csv')
df_poverty = pd.read_csv('df_ poverty.csv')
df_Trade = pd.read_csv('df_ Trade.csv')
df_agriculture = pd.read_csv('df_agriculture.csv')
df_climate = pd.read_csv('df_climate.csv')
df_Economic_Growth = pd.read_csv('df_Economic Growth.csv')
df_energy_and_extractive = pd.read_csv('df_energy_and_extractive.csv')
df_enviroment_natural_rsources = pd.read_csv('df_enviroment_natural_rsources.csv')
df_Financial_sector_development = pd.read_csv('df_Financial_sector_development.csv')
df_Health_nutrition_population = pd.read_csv('df_Health_nutrition_population.csv')
df_private_sector_development = pd.read_csv('df_private_sector_development.csv')
df_public_sector_management = pd.read_csv('df_public_sector_management.csv')
df_Social_Development = pd.read_csv('df_Social Development.csv')
df_Social_Protection = pd.read_csv('df_Social Protection.csv')
```

```python
df_education.head()
```

```python
# Drop the first column, 'Unnamed: 0'
df_education = df_education.drop(df_education.columns[0], axis=1)
df_Macroeconomic_vulnerability_debt = df_Macroeconomic_vulnerability_debt.drop(df_Macroeconomic_vulnerability_debt.columns[0], axis=1)
df_poverty = df_poverty.drop(df_poverty.columns[0], axis=1)
df_Trade = df_Trade.drop(df_Trade.columns[0], axis=1)
df_agriculture = df_agriculture.drop(df_agriculture.columns[0], axis=1)
df_climate = df_climate.drop(df_climate.columns[0], axis=1)
df_Economic_Growth = df_Economic_Growth.drop(df_Economic_Growth.columns[0], axis=1)
df_energy_and_extractive = df_energy_and_extractive.drop(df_energy_and_extractive.columns[0], axis=1)
df_enviroment_natural_rsources = df_enviroment_natural_rsources.drop(df_enviroment_natural_rsources.columns[0], axis=1)
df_Financial_sector_development = df_Financial_sector_development.drop(df_Financial_sector_development.columns[0], axis=1)
df_Health_nutrition_population = df_Health_nutrition_population.drop(df_Health_nutrition_population.columns[0], axis=1)
df_private_sector_development = df_private_sector_development.drop(df_private_sector_development.columns[0], axis=1)
df_public_sector_management = df_public_sector_management.drop(df_public_sector_management.columns[0], axis=1)
df_Social_Development = df_Social_Development.drop(df_Social_Development.columns[0], axis=1)
df_Social_Protection = df_Social_Protection.drop(df_Social_Protection.columns[0], axis=1)
```

```python
df = df_education.merge(df_Macroeconomic_vulnerability_debt, on=['Country Code', 'Country Name', 'Year'], how='inner')
df = df.drop_duplicates()
category_list_2 = [df_poverty,df_Trade,df_agriculture,df_climate,df_Economic_Growth,df_energy_and_extractive,df_enviroment_natural_rsources,df_Financial_sector_development,df_Health_nutrition_population,df_private_sector_development,df_public_sector_management,df_Social_Development,df_Social_Protection]

for i in category_list_2:
    df = df.merge(i, on=['Country Code', 'Country Name', 'Year'], how='outer')
    df = df.drop_duplicates()
```

### Dropping based on the percent values

```python
missing = pd.DataFrame(df.isna().sum()*100/df.shape[0]).reset_index()
missing = missing.rename(columns={"index": "col_name", 0: "percentage"})
drop_missing = missing[missing['percentage'] > 40].col_name.to_list() #30 corresponds to 30% for a column
```

```python
df.drop(drop_missing, axis=1,inplace=True)
```

### Dropping based on correlation

```python
cor_matrix = df.corr().abs()
#print(cor_matrix)

upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))
print(upper_tri)
```

```python
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.90)] 
print(to_drop)
```

```python
df_final = df.drop(to_drop, axis=1)
df_final.head()
```
