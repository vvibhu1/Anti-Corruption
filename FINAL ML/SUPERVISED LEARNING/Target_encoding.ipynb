```python
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('/Users/vibhuverma/Desktop/CLASSWORK/BUSINESS PRACTICUM/Anti-Corruption/Final datasets/World_Bank_development_indicators/COUNTRY LEVEL/wdi_country_top30.csv')
df.drop(['Unnamed: 0'],axis=1,inplace = True)
df.head()
df.sort_values(by=['Country Name', 'Year'],inplace=True)
```

```python
columns = df.columns.to_list()
indicators = ['Gni Growth (Annual %)',
 'Claims On Central Government (Annual Growth As % Of Broad Money)',
 'Claims On Private Sector (Annual Growth As % Of Broad Money)',
 'Inflation, Gdp Deflator: Linked Series (Annual %)',
 'Foreign Direct Investment, Net Inflows (% Of Gdp)',
 'Broad Money Growth (Annual %)',
 'Foreign Direct Investment, Net Outflows (% Of Gdp)',
 'Intentional Homicides (Per 100,000 People)',
 'Food Production Index (2004-2006 = 100)',
 'Inflation, Consumer Prices (Annual %)',
 'Net Barter Terms Of Trade Index (2000 = 100)',
 'Exports Of Goods And Services (% Of Gdp)',
 'Price Level Ratio Of Ppp Conversion Factor (Gdp) To Market Exchange Rate',
 'Total Tax And Contribution Rate (% Of Profit)',
 'Bank Liquid Reserves To Bank Assets Ratio (%)',
 'Total Reserves (% Of Total External Debt)',
 'Gross Savings (% Of Gdp)',
 'Merchandise Exports To Low- And Middle-Income Economies Within Region (% Of Total Merchandise Exports)',
 'Merchandise Exports To Low- And Middle-Income Economies Outside Region (% Of Total Merchandise Exports)',
 'Domestic Private Health Expenditure (% Of Current Health Expenditure)',
 'Agg_Index']
```

```python
for i in indicators:
    col_name = str(i)+"_MA(3)"
    df[col_name] = df.groupby('Country Name')[i].transform(lambda x: x.rolling(3).mean())   
```

```python
years = [2007, 2010, 2013, 2016, 2019]

df_years= df[df.Year.isin(years)]
```

```python
extras = ['Country Code','Corruption Perception Index','IncomeGroup','region_USAID','region_WorldBank']
drop = indicators + extras

df_years.loc[df['Year'] == 2007, 'Year'] = '2005-2007'
df_years.loc[df['Year'] == 2010, 'Year'] = '2008-2010'
df_years.loc[df['Year'] == 2013, 'Year'] = '2011-2013'
df_years.loc[df['Year'] == 2016, 'Year'] = '2014-2016'
df_years.loc[df['Year'] == 2019, 'Year'] = '2017-2019'
```

```python
df_years.drop(drop,axis=1,inplace=True)
```

```python
drop = df_years.columns.to_list()
drop.remove('Country Name')
drop.remove('Year')


for i in drop:
    col_name = str(i)+"_diff"
    df_years[col_name] = df_years.sort_values(['Country Name','Year']).groupby('Country Name')[i].diff()
```

```python
df_years.drop(drop,axis=1,inplace = True)
```
