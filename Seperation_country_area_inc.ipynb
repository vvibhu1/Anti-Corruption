```python
# Import Metadata_Country_API
df2 = pd.read_csv("Metadata_Country_API.csv")
df2.head()
```

```python
df2.shape
```

```python
# Check the number of null values
df2.isnull().sum()
```

```python
len(df2['TableName'].unique())
```

```python
# Since all the real countries belong to certain regions, and the other country codes don't have Region value. 
# So we can filter the real countries by removing null region values.
df3 = df2[['Country Code','Region','TableName']]
df3 = df3.dropna()
df3.shape
```

```python
ctr_list = df3['Country Code']
ctr_list = ctr_list.to_list()
ctr_list[:10]
```

```python
# Check the income group country code and manually add them in a list
inc_list = ['LIC','LMC','LMY','MIC','HIC','UMC']
```

```python
# Then the rest of the codes should be area code
a = all_code.tolist()
area_list = list(set(a) - set(inc_list) - set(ctr_list))
```

```python
len(area_list)
```

```python
df2['Region'].unique()
# So even though there are 41 area avaliable, only 7 of them appear in the dataset
```

```python
# Check the country rows
ctr_rows = df2.loc[df2['Country Code'].isin(ctr_list)]
ctr_rows
```

```python
# Check the rows with area code
area_rows = df2.loc[df2['Country Code'].isin(area_list)]
area_rows
```

```python
# Check the rows with income group code
inc_rows = df2.loc[df2['Country Code'].isin(inc_list)]
inc_rows
```

```python
# Since ctr_rows already has Income Group and Region information for each country, we can merge it with df1 to get the complete information.
# df4 only includes real countries with each country's income group and region. All the regions and income groups have been filtered out
df4 = df1.merge(ctr_rows, on=['Country Code'], how='inner')
df4 = df4.drop(['SpecialNotes','TableName'], axis = 1) 
df4.head()
```

```python
df4.shape
```

```python
len(df4['Country Code'].unique())
```

```python
# Similarly, create the table for region information
df5 = df1.merge(area_rows, on=['Country Code'], how='inner')
#df5 = df5.drop(['SpecialNotes','TableName'], axis = 1) 
df5.head()
```

```python
df5.shape
```

```python
# Similarly, create the table for income group information
df6 = df1.merge(inc_rows, on=['Country Code'], how='inner')
#df6 = df6.drop(['SpecialNotes','TableName'], axis = 1) 
df6.head()
```

```python
df6.shape
```

```python
# Export the three merged data tables to csv files
df4.to_csv('Ind_country.csv')
df5.to_csv('Ind_area.csv')
df6.to_csv('Ind_income.csv')
```
