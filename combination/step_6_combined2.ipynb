```python
import pandas as pd
df=pd.read_csv("df_combined.csv")
df
```

```python
df1=pd.read_csv("selected_indicators.csv")
df1.drop(['Unnamed: 0'],axis=1,inplace=True)
df1.rename(columns={'Country Name': 'country', 'Year': 'st_date'},inplace=True)
df1
```

```python
left_merged = pd.merge(df,df1, how="left", on=["country", "st_date"])
left_merged
```

```python
left_merged.drop_duplicates(subset ="uuid",inplace = True) 
left_merged
```

```python
left_merged.to_csv("final_combined.csv")
```
