```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('combined_usaid_Nov_18.csv')
```

```python
df.columns
```

```python
len(df)
```

##### we find that the values of some rows of end_date column are "-",so we delete them

```python
df=df.drop(index=(df.loc[(df['st_date']=='-')].index))
df=df.drop(index=(df.loc[(df['end_date']=='-')].index))
```

```python
len(df)
```

```python
df["end_date"]=df["end_date"].astype('int')
```

```python
df["st_date"]=df["st_date"].astype('int')
```

```python
df["duration"]=df["end_date"]-df["st_date"]
```

```python
df.columns
```

```python
df.head(5)
```

# group by contry

### duration(end_year-st_year)

```python
duration_mean=df.groupby('country')['duration'].mean()
duration_min=df.groupby('country')['duration'].min()
duration_max=df.groupby('country')['duration'].max()
duration_median=df.groupby('country')['duration'].median()
```

```python
duration_mean
```

```python
duration_mean=pd.DataFrame(duration_mean)
duration_min=pd.DataFrame(duration_min)
duration_max=pd.DataFrame(duration_max)
duration_median=pd.DataFrame(duration_median)
```

```python
duration=pd.merge(duration_mean,duration_min,on='country')
```

```python
duration
```

```python
duration=pd.merge(duration,duration_max,on='country')
```

```python
duration=pd.merge(duration,duration_median,on='country')
```

```python
columns = [ 'duration_mean', 'duration_min','duration_max','duration_median']   #改列名
duration.columns = columns
duration.columns
```

```python
duration
```
