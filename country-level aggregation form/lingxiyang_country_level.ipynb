```python
dominant_topic_4_count=df[df['dominant_topic'] == 4 ].groupby('country')['dominant_topic'].count()
dominant_topic_4_count
```

```python
import functools

dfs = [df_1,dominant_topic_4_count,dominant_topic_3_count,dominant_topic_2_count,dominant_topic_1_count,dominant_topic_0_count]

df_dominant_topic = functools.reduce(lambda left,right: pd.merge(left,right,on=["country"], how="left"), dfs).fillna(0)
df_dominant_topic
```
